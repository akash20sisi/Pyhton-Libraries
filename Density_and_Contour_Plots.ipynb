```python
%matplotlib inline
import matplotlib.pyplot as plt
plt.style.use("seaborn-v0_8-whitegrid")
import numpy as np
```

# **Visualizing a Three-Dimensional Function**

```python
def f(x, y):
 return np.sin(x) ** 10 + np.cos(10 + y * x) * np.cos(x)
```

```python
x = np.linspace(0, 5, 50)
y = np.linspace(0, 5,40)
```

```python
X, Y = np.meshgrid(x, y)
Z = f(X, Y)
```

```python
plt.contour(X,Y,Z)
```

```python
plt.contour(X, Y, Z, 4, cmap='RdGy');
plt.colorbar()
```

```python
plt.contourf(X, Y, Z, 20, cmap='RdGy')
plt.colorbar();
```

```python
contours = plt.contour(X, Y, Z, 3, colors = 'k')
plt.clabel(contours, inline=True, fontsize=8)
```

# **Histograms, Binnings, and Density**

```python
rng = np.random.default_rng(1701)
data = rng.normal(size=1000)
```

```python
plt.hist(data)
```

```python
plt.hist(data,bins =9,density = True, histtype='stepfilled', color = 'steelblue',edgecolor= 'red');
```

```python
x1 = rng.normal(0, 0.8, 1000)
x2 = rng.normal(-2, 1, 1000)
x3 = rng.normal(3, 2, 1000)
kwargs = dict(histtype='stepfilled', alpha=0.8, density=True, bins=40)
plt.hist(x1, **kwargs)
plt.hist(x2, **kwargs)
plt.hist(x3, **kwargs);
```

**If you are interested in computing, but not displaying, the histogram (that is, count‐
ing the number of points in a given bin), you can use the np.histogram function:**

```python
counts, bin_edges = np.histogram(data, bins=5)
print(counts)
print(bin_edges)
```
