```python
import matplotlib as mpl
import matplotlib.pyplot as plt
import numpy as np
```

```python
x = np.linspace(0,10,1000)
plt.plot(x,np.sin(x))
plt.plot(x,np.cos(x))
```

**Adjusting the Plot: Line Colors and Styles**

```python
plt.plot(x, np.sin(x - 0), color='blue') # specify color by name
plt.plot(x, np.sin(x - 1), color='g') # short color code (rgbcmyk)
plt.plot(x, np.sin(x - 2), color='0.75') # grayscale between 0 and 1
plt.plot(x, np.sin(x - 4), color=(1.0,0.2,0.3)) # RGB tuple, values 0 to 1
```

```python
plt.plot(x,x,linestyle = 'solid')
plt.plot(x,x+1,linestyle = 'dashed')
plt.plot(x,x+2,linestyle = 'dashdot')
plt.plot(x,x+3,linestyle = 'dotted')
# For short, you can use the following codes:
plt.plot(x, x + 4, linestyle='-') # solid
plt.plot(x, x + 5, linestyle='--') # dashed
plt.plot(x, x + 6, linestyle='-.') # dashdot
plt.plot(x, x + 7, linestyle=':'); # dotted
```

**You can save some key
strokes by combining these linestyle and color codes into a single non-keyword
argument to the plt.plot function**

* These single-character color codes reflect the standard abbreviations in the RGB
(Red/Green/Blue) and CMYK (Cyan/Magenta/Yellow/blacK) color systems, com
monly used for digital color graphics.

```python
plt.plot(x, x + 4, '-g') # solid green
plt.plot(x, x + 5, '--c') # dashed cyan
plt.plot(x, x + 6, '-.k') # dashdot black
plt.plot(x, x + 7, ':r'); # dotted red
```

**Adjusting the Plot: Axes Limits**

```python
plt.plot(x,np.sin(x))
plt.xlim(-11,11)
plt.ylim(-1,1)
plt.axis('tight');
```

```python
plt.plot(x,np.cos(x))
plt.axis('equal')
```

**Labeling Plots**

```python
plt.plot(x,np.cos(x),'--r')
plt.title("A cosine curve")
plt.xlabel("x")
plt.ylabel("cos(x)")
```

```python
plt.plot(x, np.sin(x), '-g', label='sin(x)')
plt.plot(x, np.cos(x), ':b', label='cos(x)')
plt.axis('equal')
plt.legend()
plt.xlabel("x")
```
